```python
#input sent->string
#output dictionary
def sent_ind(sent):
    la=[]
    w=""
    for i,j in enumerate(sent):
        #print(i,":",j)
        if j!=" ":
            w=w+j
            if len(w)==1:
                st=i
        else:
            la.append({"word":w,"start":st,"end":i})
            w=""
    return la
```

```python
#input lw->list(list of words)
#input tag->string
#output list of words with tags
def func_replace(lw,tag):
    ct=len(lw)
    if ct==1:
        return [[lw[0],"B-"+tag]]
    elif ct==2:
        return [[lw[0],"B-"+tag],[lw[1],"L-"+tag]]
    else:
        c=[[lw[0],"B-"+tag]]
        for i in lw[1:-1]:
               c.append([i,"I-"+tag])
        c.append([lw[-1],"L-"+tag])
        return c
```

```python
#input text->string
#input entities->list of strings
#output list(list of words with tags)
def dataset_creator(text,entities):
    d=sent_ind(text)
    lw=[]
    ck=0
    it=0
    sl=[]
    for m in entities:
        sl=[]
        ck=0
        for i in d[it:]:
            if ck==0:
                if m[0]==i['start']:
                    sl.append(i['word'])
                    ck=1
                    it=it+1
                else:
                    lw.append([i['word'],'Others'])
                    it=it+1
            else:
                if m[1]!=i['end'] and m[1]+1!=i['end']  :
                    sl.append(i['word'])
                    ck=ck+1
                    it=it+1
                else :
                    sl.append(i['word'])
                    lw.extend(func_replace(sl,m[-1]))
                    ck=0
                    it=it+1
                    break
    return text_clean(lw)
```

```python
#input data->list 
#ouput list (cleaned list of strings)
def text_clean(data):
    import re
    #print(data[0])
    for i,j in enumerate(data):
            c=re.sub(r'\W+', '', j[0])
            if c.strip()=="":
                data.pop(i)
            else:
                data[i][0]=c
    return data
```
